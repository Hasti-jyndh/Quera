```python
files = ['R98', 'R99', 'R1400', 'R1401', 'U98', 'U99', 'U1400', 'U1401']

costs = ['food', 'tobacco', 'clothing', 'housing', 'furniture', 
         'healthcare', 'transportation', 'comm', 'leisure', 'food2', 'other']

total_costs = []

for file in files:
    sheet_names = [f'{file}P3S01', f'{file}P3S02', f'{file}P3S03', f'{file}P3S04', 
                   f'{file}P3S05', f'{file}P3S06', f'{file}P3S07', f'{file}P3S08', f'{file}P3S09',
                   f'{file}P3S11', f'{file}P3S12']

             
    dfs = []
    for sheet_name, column_name in zip(sheet_names, costs):
        df = pd.read_excel(f'{file}.xlsx', sheet_name=sheet_name)
    
        df['cost_type'] = column_name
        dfs.append(df)
    
    combined_df = pd.concat(dfs, ignore_index=True)
    combined_df = combined_df.drop(columns=['code', 'purchased'])
    

    result_df = combined_df.pivot_table(index='Address', columns='cost_type', values='value', aggfunc='sum').reset_index()
    
    total_costs.append(result_df)

total_costs[0]['year'] = 98
total_costs[1]['year'] = 99
total_costs[2]['year'] = 1400
total_costs[3]['year'] = 1401

total_costs[4]['year'] = 98
total_costs[5]['year'] = 99
total_costs[6]['year'] = 1400
total_costs[7]['year'] = 1401
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer
import xgboost as xgb
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


xgb_model = xgb.XGBRegressor()
param_dist = {
    'n_estimators': [100, 150, 200],
    'learning_rate': [0.07, 0.08, 0.09],
    'max_depth': [7, 8, 9],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

random_search = RandomizedSearchCV(estimator = xgb_model,
                                   param_distributions = param_dist,
                                   n_iter = 10,
                                   cv = 3,
                                   scoring=make_scorer(r2_score),
                                   n_jobs=-1,
                                   verbose=2,
                                   random_state=42)

random_search.fit(X_train, y_train)

print(random_search.best_params_)
```

```python
eval_set = [(X_train, y_train), (X_test, y_test)]
xgb_model.fit(X_train, y_train, eval_set=eval_set, verbose=False)

evals_result = xgb_model.evals_result()

train_r2_scores = []
test_r2_scores = []

for i in range(1, 201):
    y_train_pred = xgb_model.predict(X_train, iteration_range=(0, i))
    y_test_pred = xgb_model.predict(X_test, iteration_range=(0, i))
    
    train_r2_scores.append(r2_score(y_train, y_train_pred))
    test_r2_scores.append(r2_score(y_test, y_test_pred))

x_axis = range(0, len(evals_result['validation_0']['rmse']))

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_axis, evals_result['validation_0']['rmse'], label='Train')
plt.plot(x_axis, evals_result['validation_1']['rmse'], label='Test')
plt.title('RMSE over n_estimators')
plt.xlabel('n_estimators')
plt.ylabel('RMSE')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_axis, train_r2_scores, label='Train R2')
plt.plot(x_axis, test_r2_scores, label='Test R2')
plt.title('R2 score over n_estimators')
plt.xlabel('n_estimators')
plt.ylabel('R2')
plt.legend()

plt.tight_layout()
plt.show()
```
